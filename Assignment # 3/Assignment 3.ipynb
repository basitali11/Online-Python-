```python
# Write a program which takes 5 inputs from user for different subjectâ€™s marks, total it
# and generate mark sheet using grades ?

math = int(input("Enter the number of Math subject : "))
english = int(input("Enter the number of English subject : "))
islamiyat = int(input("Enter the number of Islamiyat subject : "))
urdu = int(input("Enter the number of Urdu subject : "))
science = int(input("Enter the number of Science subject : "))

total = math + english + islamiyat + urdu + science
print ("Total marks is = ",total)
per = (total * 100) / 500
print ("Percentage is = ",per)
if per > 100:
    print("Wrong Calculate Percentage")
elif per <= 100 and per >=90:
    print ("Grade is A+")
elif per < 90 and per >=80:
    print ("Grade is A") 
elif per < 80 and per >=70:
    print ("Grade is B")
elif per < 70 and per >=60:
    print ("Grade is C")
elif per < 60 and per >=50:
    print ("Grade is D")
elif per < 50 and per >=40:
    print ("Grade is E")
else:
    print ("You are Fail")
```
