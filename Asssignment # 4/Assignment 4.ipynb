```python
# Make a calculator using Python with addition , subtraction ,
# multiplication ,division and power.

num1 = int(input("Enter the First number "))
num2 = int(input("Enter the Second number "))
op = input("Enter the Operation +,-,*,/,** ")
if op=='+':
    print("The Sum is: ",num1+num2)
elif op=='-':
    print("The Subtraction is: ",num1-num2)    
elif op=='*':
    print("The Multiply is: ",num1*num2)    
elif op=='/':
    print("The Division is: ",num1/num2)
elif op=='**':
    print("The Division is: ",num1**num2)
```

```python
# Write a program to check if there is any numeric value in list using for
# loop.

l1 = ["book", "red", 4, "car"] 
for i in l1: 
    if type(i) == int: 
        print(i)
```

```python
# Write a Python script to add a key to a dictionary.

dict = {'key1':'Book', 'key2':'Red'}
print("Current Dictionary is: ", dict)

dict.update({'key3':'Tomato'})
print("Add/Update Dictionary is: ", dict)


dict1 = {'key4':'Good Bye', 'key5':'Best Pen'}
dict.update(dict1)
print(dict)
```

```python
# Write a Python program to sum all the numeric items in a dictionary.

d={'A':10,'B':40,'C':60}
print("Total sum of values in the dictionary is : ", sum(d.values()))
```

```python
# Write a program to identify duplicate values from list.

l=[1,2,3,4,5,2,3,4,7,9,5]
l1=[]
for i in l:
    if i not in l1:
        l1.append(i)
    else:
        print(i,end=' ')
```

```python
# Write a Python script to check if a given key already exists in a
# dictionary

d = {1: 10, 2: 20, 3: 30, 4: 40, 5: 50, 6: 60}
def present(x):
    if x in d:
        print('Key is present in the dictionary')
    else:
        print('Key is not present in the dictionary')
present(5)
present(7)
```
